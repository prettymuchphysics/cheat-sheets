```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('science')
#plt.style.use(['science', 'dark_background'])
np.random.seed(3)

t = np.linspace(0,2*np.pi,500)
x, y = np.sin(t), np.cos(t)
data1 = [10, 20, 40, 80]
data2 = [12, 18, 48, 66]
data3 = np.random.randn(300)
data4 = np.random.randn(300) * 2
data5 = np.array([[1,2], [3,4], [5,6]])
labels = ['A', 'B', 'C', 'D']
X, Y = np.meshgrid(t, t)
Z = (np.sin(X) + np.cos(Y))**2
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.set_xlim(t[0], t[-1])
ax.set_ylim(-1.2, 1.2)
ax.set_xlabel(r'time')
ax.set_ylabel(r'position')
ax.plot(t, x, c='r', label=r'one')
ax.fill_between(t, x, y, alpha=0.1, color='g')
ax.plot(t, y, c='b', label=r'two')
ax.legend(title='Title')
ax.set_title(r'Oscillation of $f(x)$')
ax.annotate(r'$\swarrow$', (3.4,0))
plt.savefig('fig-single.pdf')
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.pie(data1, 
       explode=[0,0,.08,0], 
       labels=labels, 
       autopct='%1.1f\%%',
       startangle=90, 
       counterclock=False
       )
ax.axis('equal')
plt.savefig('fig-pie-chart.pdf')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(5,3))
ax = axes[0]
ax.plot(t, x, t, y, linestyle=':')
ax = axes[1]
ax.plot(t, x+y, t, x-y)
plt.savefig('fig-multi.pdf')
```

```python
fig, ax = plt.subplots(figsize=(5,2))
ax.set_xscale('log')
ax.plot(t, x, t, y)
plt.savefig('fig-logscale.pdf')
```

```python
l = np.arange(len(labels))
w = 0.35

fig, ax = plt.subplots(figsize=(5,3))
ax.bar(l - w/2, data1, w, label='a')
ax.bar(l + w/2, data2, w, label='b')
ax.set_xticks(l)
ax.set_xticklabels(labels)
ax.legend()
plt.savefig('fig-bar.pdf')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
c = ax.contour(X, Y, Z, levels=6)
ax.clabel(c, inline=True, fmt='%1.1f')
plt.savefig('fig-contour.pdf')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
c = ax.contourf(X, Y, Z, levels=6)
cbar = fig.colorbar(c)
cbar.ax.set_ylabel('value')
plt.savefig('fig-contourf.pdf')
```

```python
from matplotlib import cm
fig, axes = plt.subplots(1, 2, subplot_kw={"projection": "3d"}, figsize=(8,4))
ax = axes[0]
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
ax = axes[1]
ax.plot_surface(X, Y, Z, alpha=0.3)
ax.contour(X, Y, Z, zdir='x', offset=X.min(), cmap=cm.coolwarm)
ax.contour(X, Y, Z, zdir='y', offset=Y.max(), cmap=cm.coolwarm)
ax.contour(X, Y, Z, zdir='z', offset=Z.min(), cmap=cm.coolwarm)
plt.savefig('fig-3d-surface.pdf', transparent=True)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(data3, bins=20, density=True, histtype='step', label='a')
ax.hist(data4, bins=20, density=True, histtype='step', label='b')
ax.legend()
plt.savefig('fig-hist.pdf')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(data3, data4, marker='.')
plt.savefig('fig-scatter.pdf')
```
